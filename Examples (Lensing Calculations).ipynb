## Examples of Lensing Calculations using conv_comp.py

```python
import matplotlib.pyplot as plt
import numpy as np
import conv_comp as cc
import scipy.stats
```

### Examples

#### Snapshots 52, 63, 83

```python
mtoconv = cc.m_to_conv([52, 63, 83]) # For this example, I only have the following three files downloaded
mtoconv.dosnap2()
```

### Loading the magnification maps for each snapshot

#### Snapshot 52

First, we want to see the magnification map for this snapshot

```python
phixx_52 = np.load('phixx_maps_dictionary.npy', allow_pickle = True).item()[52]
phixy_52 = np.load('phixy_maps_dictionary.npy', allow_pickle = True).item()[52]
phiyy_52 = np.load('phiyy_maps_dictionary.npy', allow_pickle = True).item()[52]
kappa_52 = np.load('kappa_maps_dictionary.npy', allow_pickle = True).item()[52]
gammac = phixy_52
gammap = 0.5*(phixx_52 - phiyy_52)
mu = ((1 - kappa_52)**2 - gammap**2 - gammac**2)**(-1)
fig, ax = plt.subplots(figsize = (12, 10))
plt.imshow(mu,vmin =-10, vmax = 9, cmap = "RdGy", origin ='lower')
plt.colorbar();
print(f'Max mag: {np.max(mu)}')
print(f'Min mag: {np.min(mu)}')
mu_array = []
for i in range(0, 600):
    for j in range(0, 600):
        mu_array += [mu[i, j]]
```

Now, we'll look at how the magnification values are distributed. We expect the average to be close to 1 since we expect regions of overdensity to "balance" regions of underdensity. Notice that this is the distribution of the magnifications for the whole snapshot. 

```python
mean = np.mean(mu_array)
std = np.std(mu_array)
y = scipy.stats.norm.pdf(np.linspace(np.min(mu_array),np.max(mu_array),200), mean,std)
plt.plot(np.linspace(np.min(mu_array),np.max(mu_array),200), y)
print(f'mean mag:{np.mean(mu_array)}')
print(f'stdev(mag): {std}')
plt.show()
```
