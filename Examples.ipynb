## Importing the packages and file

```python
import data_ext as dext
import numpy as np
import matplotlib.pyplot as plt
import time # to check how long the new code takes
```

## Fixing the rotation matrix (Remember: This should only be run once)

```python
rotmat = dext.rotator()
```

## Example with Data

### Snapshot 52

```python
t1 = time.time()
for i in ['gas','dm', 'stars']:
    d = dext.dEx(52, 99, 49, 75000, i, rotmat)
    d.z_restrictor_bounds()
    d.data_extractor()
    d.data_reducer1()
    d.hist()
t2 = time.time()
print(f'{t2 - t1} + '' + seconds')
```

```python
D1 = np.load('density_52_dm.npy', allow_pickle = True)
D2 = np.load('density_52_gas.npy', allow_pickle = True)
D3 = np.load('density_52_stars.npy', allow_pickle = True)
```

```python
fig,ax  = plt.subplots(1,3, figsize = (22,10))
ax[0].imshow(np.log10(D1),origin = 'lower')
ax[0].set_title('Dark Matter')
ax[1].imshow(np.log10(D2),origin = 'lower')
ax[1].set_title('Gas')
ax[2].imshow(np.log10(D3),origin = 'lower')
ax[2].set_title('Stars')
```
